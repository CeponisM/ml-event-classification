```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("../data/training.csv")

# Basic info
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nMissing values:\n", df.isna().sum())

# Show a few rows
df.head()
```

```python
# Count of signal vs background
sns.countplot(data=df, x="Label")
plt.title("Signal vs Background Distribution")
plt.show()
```

```python
# Example: Plot distribution of a few features
features = ["DER_mass_MMC", "DER_mass_transverse_met_lep", "DER_pt_h"]

for feature in features:
    plt.figure(figsize=(6, 4))
    sns.histplot(data=df, x=feature, hue="Label", bins=50, kde=False, stat="density", common_norm=False)
    plt.title(f"Feature Distribution: {feature}")
    plt.show()
```
